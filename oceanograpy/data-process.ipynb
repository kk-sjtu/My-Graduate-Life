```python
import netCDF4 as nc
import numpy as np

# 打开 NetCDF 文件
file_path = '/Users/keikei/Documents/OISST/sst.day.mean.1982.nc'
nc_file = nc.Dataset(file_path)

# 获取经纬度变量
latitude = nc_file.variables['lat'][:]
longitude = nc_file.variables['lon'][:]
sst = nc_file.variables['sst']

# 指定的经纬度
target_lat = 38.15
target_lon = 114.55

# 找到最接近的经纬度索引
lat_idx = np.abs(latitude - target_lat).argmin()
lon_idx = np.abs(longitude - target_lon).argmin()

closest_lat = latitude[lat_idx]
closest_lon = longitude[lon_idx]

print(f"Closest latitude index: {lat_idx}, value: {closest_lat}")
print(f"Closest longitude index: {lon_idx}, value: {closest_lon}")

# 获取指定经纬度的 SST 数据
arr3 = sst[:, lat_idx, lon_idx]

# 将该点 SST 数据写入文件
with open('output1.txt', 'a') as file:
    for k in range(len(arr3)):
        print(arr3[k])
        file.write(f'{arr3[k]}\n')

# 关闭 NetCDF 文件
# nc_file.close()
```

```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# 打开 NetCDF 文件
file_path = '/Users/keikei/Documents/OISST/sst.day.mean.1982.nc'
nc_file = nc.Dataset(file_path, mode='r')

# 获取经纬度和 SST 变量
latitude = nc_file.variables['lat'][:]
longitude = nc_file.variables['lon'][:]
sst = nc_file.variables['sst'][0, :, :]  # 读取第一个时间步的 SST 数据

# 检查是否有掩码数据
if np.ma.is_masked(sst):
    print("Data is masked. Handling masked data.")
    sst = sst.filled(np.nan)  # 将掩码数据填充为 NaN

# 创建图像
fig = plt.figure(figsize=(10, 5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.LAKES, edgecolor='black')
ax.add_feature(cfeature.RIVERS)

# 绘制 SST 数据
sst_plot = ax.pcolormesh(longitude, latitude, sst, transform=ccrs.PlateCarree(), cmap='coolwarm')
plt.colorbar(sst_plot, ax=ax, orientation='vertical', label='SST (°C)')

# 设置标题
plt.title('Sea Surface Temperature')

# 显示图像
plt.show()

# 关闭 NetCDF 文件
# nc_file.close()
```
